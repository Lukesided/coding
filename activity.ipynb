```python
import pandas as pd

data = {'Songs': ['Alphage','Gamman','Epsiloan','Etazea','Iotanean','Lambdaha','Nuz','Omicronene','Rhozen','Tauwa','Phine','Psikick'], 'Artist': ['Alpha Beta','Gamma Delta','Alpha Beta','Eta Theta','Phi Chi','Lambda Mu','Phi Chi','Omicron Pi', 'Rho Sigma', 'Tau Upsilon', 'Phi Chi','Psi Omega'],  'Energy Level': [0.5,0.7,0.8,0.85,0.95,0.2,0.25,0.85,0.4,0.6,0.65,0.75], 'Day_Played': ['Sunday','Monday','Saturday','Saturday','Sunday','Friday','Wednesday','Saturday','Tuesday','Friday','Thursday','Friday']}
df = pd.DataFrame(data)
top_5 = df.head()
print(df)
print(top_5)
```

# **Question** **2**

```python
mode = df.mode()['Artist'][0]
print(mode)
print("The artist that I listen to the most is...", mode,"!")
```

# **Question** **3**

```python
mean = df['Energy Level'].mean()
print("My average energy level is", mean,",I guess I like high energy songs.")
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([2,1,1,1,1,3,3])
y = np.array(['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday'])
plt.title("Number of Songs per Day")
plt.barh(y,x, color = "green")
plt.show()
print("I played more music on Friday and Saturday because its the weekend, and I have more time to play music and songs.")
```
